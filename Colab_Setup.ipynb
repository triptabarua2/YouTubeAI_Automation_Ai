```python
from google.colab import files
import os

# Output ফোল্ডারের সব ভিডিও ডাউনলোড করুন
for file in os.listdir('output'):
    if file.endswith('.mp4'):
        files.download(f'output/{file}')
        print(f'✅ Downloaded: {file}')
```
